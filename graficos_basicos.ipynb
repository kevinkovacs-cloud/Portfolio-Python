# Bienvenidos! En este proyecto crearé graficos basicos.
## Primero, importamos las librerias

Usaremos numpy para las funciones matematicas y matplotlib para los graficos.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data=np.arange(0,101,10) #np.arange arroja valores de un inicio a un fin
#en este caso voy de 0 a 100, pero como Python no incluye el ultimo, voy hasta 101
#el tercer termino es un 10 porque va a ir de 10 en 10
```

```python
x_data #lo visualizamos
```

```python
y_data=x_data**2 #en este caso, será el rango previo elevado al cuadrado
```

```python
y_data #lo visualizamos
```

```python
noise=np.random.randint(-10,10,11) #Esto es para generar numeros aleatorios, respetando el rango previo
```

Con Numpy, lo que haremos ahora será añadirle este ruido (numeros aleatorios) que cree recien a y_data:

```python
y_data=np.add(y_data,noise)
```

Visualizamos lo que creamos:

```python
fig=plt.figure(1,figsize=(5,5))
plt.plot(x_data,y_data,marker='o')
```

Como observamos en el grafico, utilice marcadores. Tambien podemos cambiar el tamaño de los mismos:

```python
fig=plt.figure(1,figsize=(5,5))
plt.plot(x_data,y_data,marker='o',markersize=15)
```

```python
fig=plt.figure(1,figsize=(5,5))
plt.plot(x_data,y_data,marker='o',mfc='white',linestyle='--',label='data set 1')
plt.legend(loc='best')
```

Ahora combinare dos curvas en un mismo grafico: (notese la leve modificacion de la curva 2:

```python
fig=plt.figure(1,figsize=(5,5))
plt.plot(x_data,y_data,marker='o',mfc='white',linestyle='--',label='data set 1')
fig=plt.figure(1,figsize=(5,5))
plt.plot(x_data,y_data+500,marker='o',mfc='white',linestyle='-',label='data set 2')
plt.legend(loc='best')
```

Si bien en otro proyecto me enfocare en ecuaciones matematicas con LaTeX, en este voy a darle uso para las leyendas: (Utilizare logaritmos para que sea exponencial)

```python
fig=plt.figure(1,figsize=(5,5))
plt.semilogy(x_data,y_data,marker='o',mfc='white',linestyle='--',label='$y=x^2$')
fig=plt.figure(1,figsize=(5,5))
plt.semilogy(x_data,y_data+500,marker='o',mfc='white',linestyle='-',label='$y=x^2+500$')
plt.legend(loc='best')
```

Añadiendo los nombres a los ejes de X e Y: 

```python
fig=plt.figure(1,figsize=(5,5))
plt.semilogy(x_data,y_data,marker='o',mfc='white',linestyle='--',label='$y=x^2$')
fig=plt.figure(1,figsize=(5,5))
plt.semilogy(x_data,y_data+500,marker='o',mfc='white',linestyle='-',label='$y=x^2+500$')
plt.xlabel('variable x')
plt.ylabel('variable y')
plt.legend(loc='best')
```

# Ahora haré una demostracion basica de grafico de dispersion:

```python
x_data=np.arange(0,50,1)
y_data=np.arange(0,50,1)
colors=np.arange(0,50,1)
area=np.random.rand(50)**2
fig=plt.figure(1, figsize=(5,5))
plt.scatter(x_data,y_data,c=colors,s=area)
```

A mi parecer es muy pequeño, entonces utilizo un area mayor:

```python
x_data=np.arange(0,50,1)
y_data=np.arange(0,50,1)
colors=np.arange(0,50,1)
area=np.random.rand(50)**2*1000
fig=plt.figure(1, figsize=(5,5))
plt.scatter(x_data,y_data,c=colors,s=area,alpha=0.8)
```
